```python
# por lista de incidencia

gr = []

gr.append((1,1))
gr.append((1,2))
gr.append((1,5))
gr.append((2,3))
gr.append((2,5))
gr.append((3,4))
gr.append((5,4))
gr.append((4,6))

print(gr)
```
