```python
import os
import pathlib


path=os.path.join(str(pathlib.Path().resolve()))+"/"
```

```python
import pandas as pd


dfA = pd.read_csv(path+"Amd_Market.csv")
dfN = pd.read_csv(path+"Nvidia_Market.csv")
dfI = pd.read_csv(path+"Intel_Market.csv")
```

```python
dfA
```

```python

def transform(df):
    df['Fecha'] = pd.to_datetime(df['Fecha'], format='%m/%d/%Y')

    # Luego, cambia el formato a dd/mm/yyyy
    df['Fecha'] = df['Fecha'].dt.strftime('%d/%m/%Y')
    return df

dfN = transform(dfN)
dfA = transform(dfA)
dfI = transform(dfI)
```

```python
for col in dfN.columns:
    try:
        dfN[col] = dfN[col].str.replace('$', '').str.replace(".",",")
        dfA[col] = dfA[col].str.replace('$', '').str.replace(".",",")
        dfI[col] = dfI[col].str.replace('$', '').str.replace(".",",")
    except:
        None
```

```python
dfN
```

```python
dfA=dfA.set_index("Fecha")
dfN=dfN.set_index("Fecha")
dfI=dfI.set_index("Fecha")
```

```python
dfA.to_csv(path+"Amd_Market.csv")
dfN.to_csv(path+"Nvidia_Market.csv")
dfI.to_csv(path+"Intel_Market.csv")
```
