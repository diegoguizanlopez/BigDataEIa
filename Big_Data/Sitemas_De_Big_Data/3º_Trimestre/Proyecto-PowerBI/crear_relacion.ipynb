```python
import os
import pathlib


path=os.path.join(str(pathlib.Path().resolve()))+"/"
```

```python
import pandas as pd


dfN = pd.read_csv(path+"Nvidia_Market.csv")
dfR = pd.read_csv(path+"nvidia_reveneu.csv")
```

```python
dfR
```

```python
import numpy as np


dfY = pd.DataFrame(np.unique(pd.DatetimeIndex(dfN['Fecha']).year)).reset_index()
dfY = dfY.rename(columns={"index":"idYear",dfY.columns[1]:"Year"})
dfY
```

```python
import pandas as pd

# Asegúrate de que 'Year' en dfY y el año en 'Fecha' en dfN sean del mismo tipo
dfY['Year'] = dfY['Year'].astype(int)
dfR['Year'] = dfR['date']
dfN['Year'] = pd.DatetimeIndex(dfN['Fecha']).year

# Realiza un merge en 'Year' para combinar los datos
dfN = pd.merge(dfN, dfY[['idYear','Year']], on='Year', how='left')

dfR = pd.merge(dfR, dfY[['idYear','Year']], on='Year', how='left')
```

```python
dfY=dfY.set_index('idYear')
```

```python
dfN.to_csv(path+"Nvidia_Market.csv")
dfR.to_csv(path+"nvidia_reveneu.csv")
dfY.to_csv(path+"nvidia_market_reveneu.csv")
```
