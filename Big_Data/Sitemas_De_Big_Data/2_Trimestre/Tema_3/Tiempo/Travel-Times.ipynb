# 3.Fai un delta (diferencia) entre o TotalTime e o MovingTime. Indica cal foi o dÃ­a con maior diferencia e o de menor diferencia.

```python
import numpy as np
from pandas import timedelta_range
from tabulate import tabulate

df['Delta']=df['TotalTime']-df['MovingTime']
print("MENOR DIFERENCIA:",tabulate(df[df['Delta']==df['Delta'].min()], headers='keys', tablefmt='psql'))
print("MAYOR DIFERENCIA:",tabulate(df[df['Delta']==df['Delta'].max()], headers='keys', tablefmt='psql'))
```

# 4.Agrupa os datos por dÃ­a da semana collendo a media de tempo (o delta do apartado anterior) e indica o dÃ­a da semana de menor e maior diferencia.

```python
agrupado=pd.DataFrame(df.groupby('DayOfWeek')['Delta'].mean())
print("MENOR DIFERENCIA MEDIA:",tabulate(agrupado[agrupado['Delta']==agrupado['Delta'].min()], headers='keys', tablefmt='psql'))
print("MAYOR DIFERENCIA MEDIA:",tabulate(agrupado[agrupado['Delta']==agrupado['Delta'].max()], headers='keys', tablefmt='psql'))
```
