## Pdf de contratos de junio

Importamos el contratos junio para que recoja todas las tablas y en utf 8

```python

import pandas as pd
from tabula import read_pdf

table = read_pdf("Contratos_JUNIO.pdf", pages='all',lattice=True,encoding="utf-8")
```

Recogemos cada una de las tablas ya que hay tablas guardadas que no nos interesan

```python
start = 2
multip = 3
dfs=[]
for i in range(int(len(table)/multip)):
    #dfs.append(table[int(multip*i+start)])
    #df = table[int(multip*i+start)]
    #df.iloc
    dfs.append(table[int(multip*i+start)].iloc[:,:])

#table[2]
#table[5]
#table[8]
#table[11]
#table[14]
```

Guardamos en una lista los valores de las tablas de forma que esten una bajo otra en la misma columna y no separadas, luego le damos su nombre

```python
values=[]
for i in range(dfs[0].shape[1]):
    list_row=[]
    for df in dfs:
        list_row.append(df.iloc[:,i])
    values.append(pd.concat(list_row, ignore_index=True))


df_final = pd.DataFrame({"EXP": values[0],"CONCEPTO":values[1],"PTO. €": values[2],"PREZO €":values[3],"PRÓRROGA": values[4],"MODIF.":values[5],"PROC. ADX": values[6],"NIF":values[7],"ADXUDICATARIO": values[8],"DATA APROB.":values[9],"DATA FORMALIZ.": values[10],"N.º CONTRATO":values[11]})
```

Limpiamos los datos ya que hay bastantes errores y hay que filtrar la primera fila, eliminar los /r y -/r por separado y los que su primera fila no empiece por "PCO" y volver a Indexizar para los valores perdidos

```python
df_final = df_final[df_final['EXP'].str.startswith("PCO",na=False)]
df_final.drop(df_final.index[0],inplace=True)
df_final = df_final.replace(r'-\r', '', regex=True)
df_final = df_final.replace(r'\r', ' ', regex=True)
df_final.reset_index(drop=True, inplace=True)
```
