```python
import pathlib

path=str(pathlib.Path().resolve())+"/"
```

```python
import cv2 as cv
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
model = YOLO('yolov8n.pt')
results = model([path+"coches_aparcados.jpg"])
```

```python
r = results[0]
img_bgr = r.plot()
img = cv.cvtColor(img_bgr,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
print("BOXES")
print(r.boxes)
# XYWH da la posición de cada esquina detectada
print("_____________________")
print("KEYPOINTS")
print(r.keypoints)
print("_____________________")
print("MASKS")
print(r.masks)
print("_____________________")
print("PROBS")
print(r.probs)
print("_____________________")
print("OBB")
print(r.obb)
print("_____________________")
print("NAMES")
print(r.names)
```
