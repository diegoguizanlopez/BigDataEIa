## Dibujo de los Clusters de KMeans

```python
import numpy as np


def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids,  circle_color='w', cross_color='k'):
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=2, linewidths=12, color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    resolution=1000
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),   np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),  cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),  linewidths=1, colors='k')
    plot_data(X)
    plot_centroids(clusterer.cluster_centers_)
```

```python
X=winedata_pca[:,:2]
plot_decision_boundaries(kmeans, X)
```

```python
print(kmeans.inertia_)
print(kmeans.score(X))
```

## Analisis de KMeans

### Gráfico de inertia del modelo

Con esto podemos ver como afectan los Klusters al resultado final

```python
kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(X)
                for k in range(1, 10)]
inertias = [model.inertia_ for model in kmeans_per_k]

plt.plot(range(1, 10), inertias, "bo-")
```

### Gráfico de scores de la siluetas del modelo X Cluster

Con esto podemos ver el Score de los clusters de las siluetas con esos valores

```python
from sklearn.metrics import silhouette_score
silhouette_scores = [silhouette_score(X, model.labels_) for model in kmeans_per_k[1:]]
plt.plot(range(2, 10), silhouette_scores, "bo-")
```

### Formato de "Cuchillos" para el reflejo de Clusters a usar

Con esto podemos determinar cuantos Clusters usar teniendo en cuenta que cada "Cuchillo" tiene que llegar a la media y no ser muy diferente entre ellos

```python
import matplotlib as mpl
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter

plt.figure(figsize=(11, 9))

for k in (3, 4, 5, 6):
    plt.subplot(2, 2, k - 2)

    y_pred = kmeans_per_k[k - 1].labels_
    silhouette_coefficients = silhouette_samples(X, y_pred)

    padding = len(X) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = mpl.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (3, 5):
        plt.ylabel("Cluster")

    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)

    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title("$k={}$".format(k), fontsize=16)
```

## Entrenamiento del DBSCAN

Entrenamos el modelo con las dos primeras columnas y vemos las 10 primeras labels

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(winedata_pca[:,:2])
dbscan.labels_[:10]
```

Definimos la función de dibujo de datos del DBSCAN

```python
def plot_dbscan(dbscan, X, size):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1

    cores = dbscan.components_
    anomalies = X[anomalies_mask]

    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20, c=dbscan.labels_[core_mask])
    plt.scatter(anomalies[:, 0], anomalies[:, 1],
                c="r", marker="x", s=100)
```

Las X representan a los Outliers estimados de los datos

```python
plot_dbscan(dbscan, X, size=100)
```

Cambiamos un poco los valores para ajustarlos a más deseados usando 3 Clusters calculados anteriormente

*Las X representan Outliers de los datos*

```python
dbscan = DBSCAN(eps=0.55)
dbscan.fit(X)
plot_dbscan(dbscan, X, size=100)
```

## Probabilidad de datos X cluster

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
X_new = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])
knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

```python
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)
y_pred = dbscan.labels_[dbscan.core_sample_indices_][y_pred_idx]
y_pred[y_dist > 0.2] = -1
y_pred.ravel()
```

# Mezcla Gaussiana

Definimos la mezcla Gausiana para que use las dos primeras columnas y le indicamos que use las 2 primeras columnas, usando 3 clusters

```python

from sklearn.mixture import GaussianMixture

modelo_gmm = GaussianMixture(
                n_components    = 3,
                covariance_type = 'full',
                random_state    = 42
             )

modelo_gmm.fit(X=winedata_pca[:,:2])
```

Vemos la distribucción de los datos predecidos de las 2 primeras columnas

```python
import seaborn as sns


log_probabilidad_predicha = modelo_gmm.score_samples(X=winedata_pca[:,:2])

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 3.5))
ax.set_title('Distribución predicciones')
ax.set_xlabel('Logaritmo densidad de probabilidad');
plt.plot(log_probabilidad_predicha)
```

# Clustering Aglomerativo

Definimos un Hierachy para ver la relación del cluster en formato de árbol

```python
dendrogram = hierarchy.dendrogram(clusters)
# Distancia entre primeros clusters
plt.axhline(28.9, color='red', linestyle='--'); 
# Distancia de clusters relacionados
plt.axhline(10.75, color='crimson'); 
```
