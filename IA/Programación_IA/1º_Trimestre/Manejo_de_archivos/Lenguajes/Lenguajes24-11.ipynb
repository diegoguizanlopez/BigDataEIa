```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

filename = data_dir + "Lenguajes.csv"

df = pd.read_csv(filename)
```

```python
df=df.set_index('Language')
df
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
print(max(df['Ratings']))
colorsValue=[]
for value in df['Ratings']:
    if value == max(df['Ratings']):
        colorsValue.append('gold')
    else:
        colorsValue.append('gray')

ax.bar(df.index,df['Ratings'],color=colorsValue,label='Puntuación X Lenguaje')
ax.set_xticklabels(df.index,rotation=90)
ax.legend()
```
