## **Descenso de gradiente**

```python
# y = x^2 + 7x -3
import numpy as np
import matplotlib.pyplot as plt

def calcula_y(x):
    y = (x**2) + 7*x - 3
    return y

```

```python
x=np.array([1,2,3,4])
y = calcula_y(x)
print(f"Valor: {y}, X: {x}")
```

```python
lx = np.linspace(-15,10,200)
ly = calcula_y(lx)
```

```python
plt.plot(lx,ly,"red")
```

```python
print(f"min y = {ly.min()} indexy = {np.where(ly.min()==ly)[0]}, x = {lx[np.where(ly.min()==ly)[0]]}")
```

```python
plt.plot(lx,ly,"red")
plt.scatter(lx[np.where(ly.min()==ly)[0]],ly.min())
```

```python
plt.plot(lx,ly,"red")
x=lx[np.where(ly.min()==ly)[0]]
plt.plot([int(x),5],[ly.min(),calcula_y(5)])
#La derivada es la línea tangente al punto
```
