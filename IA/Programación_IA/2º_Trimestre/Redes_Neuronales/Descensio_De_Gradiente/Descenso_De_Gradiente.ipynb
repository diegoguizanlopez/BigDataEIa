## **Descenso de gradiente**

```python
# y = x^2 + 7x -3
import numpy as np
import matplotlib.pyplot as plt

def calcula_y(x):
    y = (x**2) + 7*x - 3
    return y

```

```python
x=np.array([1,2,3,4])
y = calcula_y(x)
print(f"Valor: {y}, X: {x}")
```

```python
lx = np.linspace(-15,10,200)
ly = calcula_y(lx)
```

```python
plt.plot(lx,ly,"red")
```

```python
print(f"min y = {ly.min()} indexy = {np.where(ly.min()==ly)[0]}, x = {lx[np.where(ly.min()==ly)[0]]}")
```

```python
plt.plot(lx,ly,"red")
plt.scatter(lx[np.where(ly.min()==ly)[0]],ly.min())
```

```python
lr = 0.1
x = -13
y = calcula_y(x)
px=[x]
py=[y]
```

```python
#Vemos que si ejecutamos varias veces nos vamos acercando a su menor en plano infinito

x_incrementada = x + 0.1
y_incrementada = calcula_y(x_incrementada) 
inc_x = x_incrementada-x
inc_y = y_incrementada-y
d = inc_y/inc_x

print(f"Incrementada x-y {x_incrementada}|{y_incrementada} valor: {inc_x}|{inc_y} d:{d}")

x = x-lr*d
y = calcula_y(x)
px.append(x)
py.append(y)
plt.plot(lx,ly,color="r")
plt.scatter(px,py)
```

```python
lr = 0.25 #Learning Rate hay que provar y del resultado decidir cual es mejor
#0.5 Es el ms exacto por lo q se usara para ver el funcionamiento dejo as
x = -13
y = calcula_y(x)
px=[x]
py=[y]
```

```python
for paso in range(50):
    x_incrementada = x + 0.1
    y_incrementada = calcula_y(x_incrementada) 
    inc_x = x_incrementada-x
    inc_y = y_incrementada-y
    d = inc_y/inc_x


    print(f"Incrementada x-y {x_incrementada}|{y_incrementada} valor: {inc_x}|{inc_y} Multiplicatorio:{d}")
    x = x-lr*d
    y = calcula_y(x)
    px.append(x)
    py.append(y)
    plt.plot(lx,ly,color="r")
    plt.scatter(px,py)
```
