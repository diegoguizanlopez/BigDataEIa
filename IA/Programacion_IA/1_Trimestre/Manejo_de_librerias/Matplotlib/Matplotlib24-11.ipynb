```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alfa = 0.7
beta = 10
x = np.linspace(0, 2*np.pi, 1000)       #100 Numeros entre y 2*PI
y = np.exp(-alfa*x) * np.sin(beta*x)    #Exponencial * el seno
fig = plt.figure(figsize=[15,10])
#plt.plot(x,y)
ax = fig.add_subplot(2,2,1)             #Nmero de filas, de columnas y donde se posiciona
ax.plot(x,y)
ax = fig.add_subplot(2,2,2)
ax.plot(2*x,y)
ax = fig.add_subplot(2,2,3)
ax.plot(0.1*x,y)
ax = fig.add_subplot(2,2,4)
ax.plot(0.5*x,y)
```

```python
fig = plt.figure(figsize=[15,10])
ax = fig.add_subplot(1,1,1)
ax.plot(x,y,"r+", label= "Probando lnea",color="green")
ax.legend()     #Texto
ax.grid()       #Lineado
ax.set_xlabel("Estas son las x")
ax.set_ylabel("Estas son las y")
```
