```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + 0.2
y2 = np.sin(40*x) + 3
y  = y1 + y2
plt.plot(x, y1+5, color="b")
plt.plot(x, y2+5, color="g")
plt.plot(x, y+5, color="r")

# Recibimos el rojo, pero lo que nos interesa es el azul por lo que habrá que filtrar
f1 = np.ones(11)
#f1[5] = 0.2
f1 = f1/f1.sum()

filtrado = []
i = 0
l1 = len(f1)
while(i < (len(x) - len(f1))):
  s1 = y[i:i+l1]
  s2 = (s1 * f1).sum()
  filtrado.append(s2)
  i += 1

# ROJO TRANSFORMADO
plt.plot(x[l1//2 : -l1//2], filtrado, color="orange")

# VERDE
#plt.plot(x[l1//2 : -l1//2], y[l1//2 : -l1//2]-filtrado, color="green")
```
