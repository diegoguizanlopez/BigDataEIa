```python
import pathlib
path=str(pathlib.Path().resolve().parent)+"/imagenes/"
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

img1 = cv.imread(path + "j0.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
plt.imshow(img1,cmap="gray")
```

```python
import numpy as np

# Creamos un array de 5x5 de ints de unos
kernel = np.ones((5,5),np.uint8)

# Erosión reduce la imagen del Kernel que situe esos 1 
erosion = cv.erode(img1,kernel,iterations=1)
fig,(ax1,ax2) = plt.subplots(1,2)

ax1.imshow(img1,cmap="gray")
ax2.imshow(erosion,cmap="gray")
```

```python
import numpy as np

# Creamos un array de 5x5 de ints de unos
kernel = np.ones((5,5),np.uint8)
# Dilate aumenta el tamaño de la propia imagen en comparación de la original
dilate = cv.dilate(img1,kernel,iterations=1)
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.imshow(img1,cmap="gray")
ax2.imshow(dilate,cmap="gray")
```

```python
img1 = cv.imread(path + "j1.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)

kernel = np.ones((5,5),np.uint8)
# Permite abrir y cerrar obligando así a que desaparezcan esos puntos blancos
opening = cv.morphologyEx(img1, cv.MORPH_OPEN, kernel)

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(img1, cmap='gray')
ax1.set_title("Org")
ax2.imshow(opening, cmap='gray')
ax2.set_title("Opening")
```

```python
img1 = cv.imread(path + "j2.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)

kernel = np.ones((5,5),np.uint8)
# Permite cerrar y abrir obligando así a que desaparezcan esos puntos negros
closing = cv.morphologyEx(img1, cv.MORPH_CLOSE, kernel)

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(img1, cmap='gray')
ax1.set_title("Org")
ax2.imshow(closing, cmap='gray')
ax2.set_title("Closing")
```

```python
img1 = cv.imread(path + "j0.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)

kernel = np.ones((5,5),np.uint8)
# Permite cerrar y abrir obligando así a que desaparezcan esos puntos negros
gradient = cv.morphologyEx(img1, cv.MORPH_GRADIENT, kernel)

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(img1, cmap='gray')
ax1.set_title("Org")
ax2.imshow(gradient, cmap='gray')
ax2.set_title("Gradient")
```

```python
img1 = cv.imread(path + "j0.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)

# Nos permite a partir de una máscara sacar su esqueleto
distanciaT = cv.distanceTransform(img1, cv.DIST_L2,0)

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(img1, cmap='gray')
ax1.set_title("Org")
ax2.imshow(distanciaT, cmap='gray')
ax2.set_title("Distancia Transform")
```

```python
img1 = cv.imread(path + "letras.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)


kernel = np.ones((5,5),np.uint8)
img2 = cv.morphologyEx(img1, cv.MORPH_CLOSE, kernel, iterations=4) # limpiamos letras
img2 = cv.morphologyEx(img2, cv.MORPH_OPEN, kernel, iterations=3) # limpiamos fondo

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(img1, cmap='gray')
ax1.set_title("Org")
ax2.imshow(img2, cmap='gray')
ax2.set_title("Close and Open")
```
