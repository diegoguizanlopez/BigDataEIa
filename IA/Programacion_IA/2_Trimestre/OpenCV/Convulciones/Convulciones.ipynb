```python
import pathlib
path=str(pathlib.Path().resolve().parent)+"/imagenes/"
```

```python
import cv2 as cv
import numpy as np

from matplotlib import pyplot as plt
src = cv.imread(path + "test.png")
src = cv.cvtColor(src, cv.COLOR_BGR2RGB)

plt.imshow(src)
```

```python
filtro = np.ones((5,5), np.float32)/25
img2 = cv.filter2D(src=src, ddepth=-1, kernel=filtro )
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(src)
ax2.imshow(img2)
```

```python
img2 = cv.blur(src=src, ksize=(5,5))
img3 = cv.GaussianBlur(src=src, ksize=(5,5), sigmaX=0, sigmaY=0 )
img4 = cv.medianBlur(src, ksize=5)
img5 = cv.bilateralFilter(src=src, d=9, sigmaColor=75, sigmaSpace=75)
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(1,5,figsize=(30,30))
ax1.imshow(src)
ax2.imshow(img2)
ax3.imshow(img3)
ax4.imshow(img4)
ax5.imshow(img5)
```
