```python
import pathlib
path=str(pathlib.Path().resolve().parent)+"/imagenes/"
```

```python
import cv2 as cv
import matplotlib.pyplot as plt


img_file = path + "ajedrez.jpg"
img = cv.imread(img_file)
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
gris = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
import numpy as np

# Lo hacemos float32 para trabajar con el
img1 = np.float32(gris)

dest = cv.cornerHarris(img1,2,5,0.07)

# CON ESTO SE VE UN POCO PUNTOS AZULES EN EL TABLERO
maximo = dest.max()
umbral = maximo * 0.05
img4 = img2.copy()
img4[dest > umbral] = [255,0,0]

plt.imshow(img4, cmap = "gray")
```

```python
img_file = path + "ajedrez.jpg"
img = cv.imread(img_file)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
esquinas = cv.goodFeaturesToTrack(gris,100,0.3,10)
esquinas = np.intp(esquinas)
for esquina in esquinas:
    x,y = esquina[0]
    cv.circle(img,(x,y),3,(255,0,0),-1)

plt.imshow(img)
```
