```python
import pathlib
path=str(pathlib.Path().resolve().parent.parent)+"/imagenes/"
```

```python
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread(path + "monedas_clase.jpg")
fig,axs = plt.subplots(1,2)
axs[0].imshow(cv.cvtColor(img,cv.COLOR_BGR2GRAY),cmap="gray")
axs[1].imshow(img[:,:,0],cmap="gray")
img1=img[:,:,0]
```

```python
ret,th1 = cv.threshold(img1,127, 255,cv.THRESH_BINARY_INV)
plt.imshow(th1,cmap="gray")
```

```python

import numpy as np

kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (7,7))


img2 = cv.morphologyEx(th1, cv.MORPH_CLOSE, kernel, iterations=2)
img2 = cv.morphologyEx(img2, cv.MORPH_OPEN, kernel, iterations=3)
plt.imshow(img2,cmap="gray")
```

```python
from cv2 import cvtColor
import numpy as np

src1 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
img = cv.cvtColor(src1, cv.COLOR_BGR2GRAY)
```
