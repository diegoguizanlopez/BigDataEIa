# **ÁRBOL V2 ORDENADOS**

```python
class Nodo:
  def __init__(self, datos):
    self.izda = None
    self.dcha = None
    self.datos = datos

class ArbolBinario:
  def __init__(self):
    self.root = None

  def insert(self, nuevo):
    if self.root is None:
      self.root = Nodo(nuevo)
    else:
      self.insert_recursivo(self.root, nuevo)

  def insert_recursivo(self, actual, nuevo):
      ret = actual.datos.compara(nuevo)
      if ret < 0:
        if actual.izda is None:
          actual.izda = Nodo(nuevo)
        else: self.insert_recursivo(actual.izda, nuevo)
      else:
        if actual.dcha is None:
          actual.dcha = Nodo(nuevo)
        else: self.insert_recursivo(actual.dcha, nuevo)

  def busca(self, clave):
    return self.busca_recursivo(self.root, clave)

  def busca_recursivo(self, actual, clave):
    if actual is None: return None
    ret = actual.datos.compara(clave)
    if ret < 0:
      return self.busca_recursivo(actual.izda, clave)
    elif ret > 0:
      return self.busca_recursivo(actual.dcha, clave)
    else: return actual.datos

  def profundidad(self):
    return self.profundidad_recursivo(self.root)

  def profundidad_recursivo(self, actual):
    if actual is None: return None
    if actual.izda: self.profundidad_recursivo(actual.izda)
    print(actual.datos)
    if actual.dcha: self.profundidad_recursivo(actual.dcha)

class Persona:
  def __init__(self, dni, nombre="Desconocido", edad = 0):
    if isinstance(dni, int): dni = str(dni)
    self.dni = dni
    self.nombre = nombre
    self.edad = edad

  def compara(self, otra_cosa):
    if isinstance(otra_cosa, int): otra_cosa = str(otra_cosa)
    if isinstance(otra_cosa, Persona): otra_cosa = str(otra_cosa.dni)
    if self.dni < otra_cosa: return 1
    if self.dni > otra_cosa: return -1
    return 0
  
  def __str__(self):
    return f"Persona: dni: {self.dni} nombre: {self.nombre} edad: {self.edad}"


dnis = ["777", "333", "999", "888", "111", "555", "444"]
t = ArbolBinario()
for d in dnis:
  p = Persona(d, f"", 0)
  t.insert(p)

t.busca("777")
```

```python
# pruebas con muchos registros
import random
import time
# una lista con 100.000 números entre 10.000.000 y 20.000.000
nelementos = 1000000
print(f"Creando {nelementos}...")
lista_grande = random.sample(range(10000000, 20000000), nelementos)
# insertamos en un árbol
print(f"Insertando...")
t = ArbolBinario()
for n in lista_grande:
  t.insert(Persona(n))

# creamos una lista de 100 elementos que existan en el árbol
```

**LECUTRA**

```python
ntest = 100
print("Creando test de ntest")
l2 = random.sample(lista_grande, ntest)
print("Test...")
t1 = time.time()
for l in l2:
  print(f"buscando {l}")
  p = t.busca(l)
  if p: print(f"Encontrado {p}")
  else: print("No encontrado")
  t2 = time.time()
print(f"Tiempo en buscar 100 elementos: {t2-t1}")
```
