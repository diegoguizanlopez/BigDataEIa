# **MATPLOTLIB**
Librería de control de gráficos

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#fig = plt.figure(figsize=(6,4))
x=[1,2,3,4]
y=[6,10,9,12]
fig,ax = plt.subplots()
ax.plot(x,y)    #Generamos un cuadro con los valores del array representados
```

```python
fig,ax = plt.subplots()
x = np.linspace(0,7,50)    #Crea un array de 0-7 con 20 valores
y = np.sin(x)              #SENO DE X
y2 = np.cos(x)             #COSENO DE X
ax.plot(x,y,label="SIN",linewidth=1,color="blue",marker="o")
ax.plot(x,y2,label="COS",linewidth=3,color="red")
ax.legend()
ax.set_title("Prueba de Matplot")
ax.set_xlim(0,2)           #Reduce el eje X
ax.set_ylim(0.5,1)
ax.annotate("SE CRUZA",xy=[0.78,0.705],xytext=[1.25,0.75],arrowprops=dict(facecolor="black"))   #Pone un texto con una flecha
ax.plot([0, 0.78], [0.705, 0.705], color="gray", linestyle="--")                                #Línea horizontal
ax.plot([0.78, 0.78], [0, 0.705], color="gray", linestyle="--")                                 #Línea vertical
ax.text(0.375,0.675,"0.705,0.78")
ax.scatter(0.78, 0.705, s=300)                                                                   #PUNTO
```
